# Web Crawler using BeautifulSoup library of python

first , import the beautifulsoup module in your script

```python
from bs4 import BeautifulSoup 
import requests
```

here we have defined maximum no of links we want to crawl through a given string via google search query

```python
max_links=200
```

```python
def crawl(start,f,recorded_links):
    if len(recorded_links)>max_links:
        print('recorded')
        return 0
    #get response from given url.
    html_stuff=requests.get(start)
    
    #if response is positive then only continue crawling through given link.
    try :
        soup=BeautifulSoup(html_stuff.text,'lxml')
        #find all of the anchor tags in the html document recieved from the url.
        links=soup.find_all('a')
        #go through each link and perform some manipulation if needed.
        for link in links:
            print(link['href'])
            mylink=''
            try:
                if link['href'][0]=='/':
                    mylink='https://www.google.com'+link['href']
                else:
                    mylink=link['href']
            #encountering index out of range exception
            except IndexError:
                mylink=''
            #if the link is already recorded than don't crawl through the link.
            if mylink not in recorded_links:
                #most of the links with shorter lengths are of no use.
                if len(mylink)>100:
                    print(mylink)
                    recorded_links.append(mylink)
                    f.write(mylink) #writing link to the file
                    f.write('\n')
                    #crawling through present link as seed
                    crawl(mylink,f,recorded_links) 
    #if response is not positive
    except HTTPError as e:
        if e.response.status_code == 404:
            return "0"
            
```

```python
def get_input():
    #get the search query from user
    start=input("Enter the search Query")
    f=open('links.txt','w+')
    google_link='https://www.google.com/search?q='
    words=start.split(' ')
    for word in words:
        google_link+=word+'+'
    google_link=google_link[:len(google_link)] 
    #prepare link for a google search
    google_link+='&aqs=chrome..69i57j0j69i60j69i61l2j0.873j0j7&sourceid=chrome&ie=UTF-8'
    print(google_link)
    crawl(google_link,f,[])
    f.close()
```
